```python
#function to find the universal(ultimate) structure of the nested dictionary
# import pandas as pd
import pprint

def get_all_keys(input_dict, path=None):
    """
    Recursively extracts all key paths from a nested dictionary.
    """
    if path is None:
        path = []
    keys = []
    for key, value in input_dict.items():
        current_path = path + [key]
        keys.append(tuple(current_path))
        if isinstance(value, dict):
            keys.extend(get_all_keys(value, current_path))
    return keys

def build_universal_structure(all_keys):
    """
    Builds a universal dictionary structure from a list of key paths.
    """
    universal_structure = {}
    for key_path in all_keys:
        current_level = universal_structure
        for i, key in enumerate(key_path):
            if key not in current_level:
                current_level[key] = {}
            if i < len(key_path) - 1:
                current_level = current_level[key]
            # If it's the last key, we've reached the end of this path
    return universal_structure

def find_universal_dict_structure(df, column_name):
    """
    Finds the universal nested dictionary structure in a DataFrame column.

    Args:
        df (pd.DataFrame): The DataFrame.
        column_name (str): The name of the column containing dictionaries.

    Returns:
        dict or None: The universal dictionary structure found, or None if no dictionaries are found.
    """
    all_found_keys = []
    for index, row in df.iterrows():
        dict_value = row[column_name]
        if isinstance(dict_value, dict):
            all_found_keys.extend(get_all_keys(dict_value))

    if not all_found_keys:
        return None

    universal_structure = build_universal_structure(all_found_keys)
    return universal_structure

# # Example Usage:
# data = {'nested_data': [
#     {'a': 1, 'b': {'c': 2, 'd': 3}},
#     {'a': 4, 'b': {'c': 5}},
#     {'a': 6, 'e': 7, 'f': {'g': 8, 'h': 9, 'i': 10}},
#     {'a': 11},
#     {'a': 12, 'b': {'c': 13, 'd': 14}},
#     {'f': {'g': 15, 'h': 16}}
# ]}
# df = pd.DataFrame(data)

# universal_structure = find_universal_dict_structure(df, 'nested_data')

# if universal_structure:
#     print("Universal Dictionary Structure:")
#     pprint.pprint(universal_structure)
# else:
#     print("No dictionaries found in the specified column.")
```

```python
# Standardiesd the year
import re
import numpy as np

def clean_year(value):
    if pd.isnull(value):
        return None
    # Extract only the first valid 4-digit year
    match = re.search(r'(18|19|20)\d{2}', str(value))
    if match:
        return int(match.group(0))
    return None

movies_df['year_raw'] = movies_df['year']  # Preserve the original
movies_df['year'] = movies_df['year_raw'].apply(clean_year)
```
